```python
import string
```

```python
#DICT = ''' \\0123456789abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ!"$%&'()*+,-./:;<=>?@[]^_`{|}~#\n\t'''
DICT = string.printable
DICT_LEN = len(DICT)
```

```python
class Encrypt():
    
    ''' This class is a collection of all methods used to encrypt text. ''' 
        
    @staticmethod
    def get_text():
        
        ''' This function is to get the text that is to be encrypted. '''
        text = str(input('Enter the text to be encrypted : '))
        return text
    
    
    @staticmethod    
    def get_text_from_file():
        
        ''' This function is to get the text from a file that is to be encrypted. '''        
        file = str(input('''Enter the file with path to be encrypted :  '''))
        try:
            with open(file,'r') as f:
                f_contents = f.read()
        except FileNotFoundError:
            print('File not found !')
        else:   
            #print(f_contents)
            return f_contents,file
    
    
    @staticmethod
    def get_key():
        
        ### DICT_KEYS are the map of original letters 
        DICT_KEYS = dict()
        
        ### Input ###
        while True:
            try:
                en_key = int(input('Enter the key : '))
                if type(en_key) == type(float()):
                    raise ValueError
            except ValueError:
                print(f'Key can only be a positive or negative integer number!')
            else:
                break
        
        original_key = en_key
        
        ### TURN THE LETTERS ###
        if en_key == 0:       # Key is 0   
           
            DICT_KEYS = { letter:letter for letter in DICT }
        
        else:                      # Key is positive or negative
            
            en_key %= DICT_LEN   #If absolute value of key greater DICT_LEN it will reduce the key else will remain same 
            for index,letter in enumerate(DICT):

                new_index = (en_key  + index) % DICT_LEN   # since new_index will go over dict_len , cycle  
                DICT_KEYS[letter] = DICT[new_index]        # Creates dictionary map

        return DICT_KEYS, original_key, en_key                   
    
    
    @staticmethod
    def encrypt_text(text, dict_keys, original_key, new_key):
        
        return ''.join([ dict_keys[letter] for letter in text ]), original_key, new_key

    
    @staticmethod
    def encrypt_text_to_file(text, file, dict_keys, original_key, new_key):
        
        print(file)
        new_file = file + '.caeser'
        text = ''.join([ dict_keys[letter] for letter in text ])
        with open(new_file,'w') as f:
            f.write(text)
        

    
```

```python
class Decrypt():
    
    ''' This class is a collection of all methods used to encrypt text. ''' 
        
    @staticmethod
    def get_text():
        text = str(input('Enter the text to be Decrypted : '))
        return text
    
    
    @staticmethod    
    def get_text_from_file():
        

        file = str(input('''Enter the file with path to be decrypted :  '''))
        try:
            with open(file,'r') as f:
                f_contents = f.read()
        except FileNotFoundError:
            print('File not found !')
        else:   
            #print(f_contents)
            return f_contents,file   
    
    
    @staticmethod
    def get_key():
        
        DICT_KEYS = dict()
        
        ### Input ###
        while True:
            try:
                en_key = int(input('Enter the key [ old or shortened ] : '))
                if type(en_key) == type(float()):
                    raise ValueError
            except ValueError:
                print(f'Key can only be a positive or negative integer number!')
            else:
                break
        
        original_key = en_key
        
        ### TURN THE LETTERS ###
        if en_key == 0:       # Key is 0   
           
            DICT_KEYS = { letter:letter for letter in DICT }
        
        else:                      # Key is positive or negative
            
            en_key %= DICT_LEN      #If absolute value of key greater DICT_LEN it will reduce the key else will remain same 
            en_key = -(en_key) #Decrypt means , negative key 
            for index,letter in enumerate(DICT):

                #new_index = abs( index - en_key ) % DICT_LEN   # since new_index will go over dict_len , cycle  
                new_index = (en_key  + index) % DICT_LEN
                DICT_KEYS[letter] = DICT[new_index]            # Creates dictionary map

        
        #print(DICT_KEYS)
        return DICT_KEYS, original_key, abs(en_key)                   
    
    
    @staticmethod
    def decrypt_text(text, dict_keys, original_key, new_key):
        
        return ''.join([ dict_keys[letter] for letter in text ]), original_key, new_key
   
    @staticmethod
    def decrypt_text_to_file(text, file, dict_keys, original_key, new_key):
        
        print(file)
        new_file = file + '.caeser_decrypt'
        text = ''.join([ dict_keys[letter] for letter in text ])
        with open(new_file,'w') as f:
            f.write(text)
```

```python
if __name__ == "__main__":
    
    x = str(input('E or d:'))
    y = str(input('F or t:'))
    x = x.lower()
    y = y.lower()
    if x == 'e' and y == 't':
        e = Encrypt() 
        text = e.get_text()
        dict_keys, original_key, new_key = e.get_key()
        for index,i in enumerate(e.encrypt_text(text, dict_keys, original_key, new_key)):
            if index == 0:
                print(f'''Encrypted text : '{i}' ''')
            elif index == 1:
                print(f'''Old key : '{i}' ''')  
            else:
                print(f'''Shortened key : '{i}' ''')  
    elif x == 'd' and y == 't':
        d = Decrypt() 
        text = d.get_text()
        dict_keys, original_key, new_key = d.get_key()
        for index,i in enumerate(d.decrypt_text(text, dict_keys, original_key, new_key)):
            if index == 0:
                print(f'''Encrypted text : '{i}' ''')
            elif index == 1:
                print(f'''Old key : '{i}' ''')  
            else:
                print(f'''Shortened key : '{i}' ''')      
    elif x == 'e' and y == 'f':
        e = Encrypt() 
        text,file = e.get_text_from_file()
        dict_keys, original_key, new_key = e.get_key()
        e.encrypt_text_to_file(text, file, dict_keys, original_key, new_key) 
    elif x == 'd' and y == 'f':
        d = Decrypt() 
        text,file = d.get_text_from_file()
        dict_keys, original_key, new_key = d.get_key()
        d.decrypt_text_to_file(text, file, dict_keys, original_key, new_key)

    
        
    
    
    
```
